2265. Count Nodes Equal to Average of Subtree
Given the root of a binary tree, return the number of nodes where the value of the node is equal to the average of the values in its subtree.

Note:

The average of n elements is the sum of the n elements divided by n and rounded down to the nearest integer.
A subtree of root is a tree consisting of root and all of its descendants



Input: root = [4,8,5,0,1,null,6]
Output: 5
Explanation: 
For the node with value 4: The average of its subtree is (4 + 8 + 5 + 0 + 1 + 6) / 6 = 24 / 6 = 4.

For the node with value 5: The average of its subtree is (5 + 6) / 2 = 11 / 2 = 5.

For the node with value 0: The average of its subtree is 0 / 1 = 0.

For the node with value 1: The average of its subtree is 1 / 1 = 1.

For the node with value 6: The average of its subtree is 6 / 1 = 6.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def averageOfSubtree(self, root: TreeNode) -> int:
        self.count = 0
        self.dfs(root)
        return self.count
    def dfs(self,node):
        if not node:
            return 0,0
        if not node.left and not node.right:
            self.count+=1
            return node.val,1
        left_sum,left_count = self.dfs(node.left)
        right_sum,right_count = self.dfs(node.right)
        average = (left_sum+right_sum+node.val)//(left_count+right_count+1)
        if average == node.val:
            self.count+=1
        return left_sum+right_sum+node.val,left_count+right_count+1
    
if __name__ == "__main__":
    sol = Solution()
    root = TreeNode(1)
    node2 = TreeNode(2)
    node3 = TreeNode(3)
    node5 = TreeNode(5)
    node4 = TreeNode(4)

    # Connect the nodes according to the tree structure
    root.left = node2
    root.right = node3
    node2.right = node5
    node3.right = node4
    val = sol.averageOfSubtree(root)
    print(val)
```
